```python
# 1. imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import validation_curve
```

```python
#load the dataset using pandas
columns = ['buying_price', 'maint_cost', 'doors', 'person_capacity', 
           'lug_boot', 'safety', 'class']

df = pd.read_csv('car_data.csv', names=columns)
```

```python
df.head()
```

```python
# 2 One-hot enconding using pandas
# 2.1 Separate features and target
X = df.drop('class', axis=1)
y = df['class']

# One-hot encode all categorical features
X_encoded = pd.get_dummies(X, columns=X.columns)
```

```python
# 3 split the data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# 4 Feature Scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
# Transform test data using the training data parameters
X_test_scaled = scaler.transform(X_test)

# Convert back to DataFrames to keep column names
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

# Summary Graphs

```python
#plot validation curve
def plot_validation_curve_rf(X, y):
    """Simple plot showing Random Forest performance with different numbers of trees"""
    n_estimators = [1, 10, 50, 100, 200]  # fewer points, easier to read
    train_scores = []
    test_scores = []
    
    for n in n_estimators:
        rf = RandomForestClassifier(n_estimators=n, random_state=42)
        rf.fit(X, y)
        train_scores.append(rf.score(X, y))
        test_scores.append(rf.score(X_test, y_test))
    
    plt.figure(figsize=(8, 6))
    plt.plot(n_estimators, train_scores, label='Training score', marker='o')
    plt.plot(n_estimators, test_scores, label='Test score', marker='o')
    plt.xlabel('Number of Trees')
    plt.ylabel('Accuracy Score')
    plt.title('Random Forest Performance vs Number of Trees')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#plot feature distribution
def plot_feature_distributions(df):
    """
    Create bar plots for each feature showing the distribution of values
    """
    # Get features (exclude the target 'class' column)
    features = df.columns[:-1]
    
    # Calculate number of rows needed (2 plots per row)
    n_rows = (len(features) + 1) // 2
    
    # Create subplots
    plt.figure(figsize=(15, 5*n_rows))
    
    # Plot each feature
    for idx, feature in enumerate(features, 1):
        plt.subplot(n_rows, 2, idx)
        sns.countplot(data=df, x=feature)
        plt.title(f'Distribution of {feature}')
        plt.xticks(rotation=45)  # Rotate labels for better readability
    
    plt.tight_layout()
    plt.show()
```

```python
#plot model accuracies
def plot_model_accuracies(dt_acc, rf_acc):
    # Create data for plotting
    models = ['Decision Tree', 'Random Forest']
    accuracies = [dt_acc, rf_acc]
    
    # Create bar plot
    plt.figure(figsize=(8, 6))
    bars = plt.bar(models, accuracies)
    
    # Customize the plot
    plt.title('Model Accuracy Comparison')
    plt.ylabel('Accuracy')
    
    # Add value labels on top of each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.4f}',
                ha='center', va='bottom')
    
    # Set y-axis to start from 0
    plt.ylim(0, 1.1)
    
    plt.show()
```

```python
#plot confusion matrices
def plot_confusion_matrices(dt_matrix, rf_matrix):
    """Plot confusion matrices for both models side by side"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # Plot Decision Tree confusion matrix
    sns.heatmap(dt_matrix, annot=True, fmt='d', cmap='Blues', ax=ax1)
    ax1.set_title('Confusion Matrix - Decision Tree')
    ax1.set_ylabel('True Label')
    ax1.set_xlabel('Predicted Label')
    
    # Plot Random Forest confusion matrix
    sns.heatmap(rf_matrix, annot=True, fmt='d', cmap='Blues', ax=ax2)
    ax2.set_title('Confusion Matrix - Random Forest')
    ax2.set_ylabel('True Label')
    ax2.set_xlabel('Predicted Label')
    
    plt.tight_layout()
    plt.show()
```

```python
#plot model metrics
def plot_model_metrics(dt_report, rf_report):
    """Plot precision, recall, and F1 scores for both models"""
    # Convert classification reports to dictionaries
    dt_dict = classification_report(y_test, dt_model.predict(X_test_scaled), output_dict=True)
    rf_dict = classification_report(y_test, rf_model.predict(X_test_scaled), output_dict=True)
    
    # Create data for plotting
    metrics = ['Precision', 'Recall', 'F1-score']
    dt_scores = [dt_dict['weighted avg'][m.lower()] for m in metrics]
    rf_scores = [rf_dict['weighted avg'][m.lower()] for m in metrics]
    
    # Set up bar positions
    x = np.arange(len(metrics))
    width = 0.35
    
    # Create plot
    plt.figure(figsize=(10, 6))
    plt.bar(x - width/2, dt_scores, width, label='Decision Tree')
    plt.bar(x + width/2, rf_scores, width, label='Random Forest')
    
    # Customize plot
    plt.ylabel('Scores')
    plt.title('Model Performance Metrics')
    plt.xticks(x, metrics)
    plt.legend()
    
    # Add value labels on bars
    for i in range(len(metrics)):
        plt.text(i - width/2, dt_scores[i], f'{dt_scores[i]:.3f}', 
                ha='center', va='bottom')
        plt.text(i + width/2, rf_scores[i], f'{rf_scores[i]:.3f}', 
                ha='center', va='bottom')
    
    plt.ylim(0, 1.1)
    plt.show()
```

# Car Evaluation Decision Tree Random Forest Report
This project analyzes a dataset of car attributes to predict their acceptability rating. We use machine learning models to classify cars into different categories (unacceptable, acceptable, good, very good) based on their features.

### 1. Initial Data Loading
The car dataset contains multiple attributes about cars, including:

Cost of buying the car
Maintenance costs
Number of doors
Passenger capacity
Luggage boot size
Safety rating
Classification (our target variable)

### 2. Data Preprocessing
First, we handle our categorical data:

Convert categorical features into numerical form using one-hot encoding

This transforms categories like 'high', 'med', 'low' into separate binary columns
Each feature gets split into multiple columns, one for each possible value


Transform the target variable (class) using label encoding

Converts categorical labels into numerical values
Preserves the ability to convert back to original labels later

### 3. Data Splitting
We divide our data into training and testing sets:

80% of data goes to training
20% is reserved for testing

### 4. Feature Scaling
Apply standard scaling to normalize our features:

Calculate mean and standard deviation from training data
Transform training data
Use same scaling parameters on test data
This ensures all features are on the same scale

### 5. Modeling Training
We train two different models:
Decision Tree Model:

Simple, interpretable model
Makes decisions based on feature thresholds
Can capture non-linear relationships

Random Forest Model:

Ensemble of multiple decision trees
Generally more robust than single decision trees
Often achieves better performance
